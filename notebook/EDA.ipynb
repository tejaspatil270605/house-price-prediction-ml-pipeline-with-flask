```python
columns = [['total_rooms','total_bedrooms','population','households','median_income']]
```

```python
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    iqr = Q3 - Q1
    lower = Q1 - 1.5 * iqr
    upper = Q3 + 1.5 * iqr 
    # df[col] = df[col].clip(lower,upper)  -- use when there are not NAN values
    df[col] = np.where(df[col]>upper,
                          upper,
                          np.where(df[col]<lower,
                                   lower,
                                   df[col]))
```

```python
plt.figure(figsize=(19,8))
sns.boxplot(df)
plt.grid(True)
```
